```python
country_names=df_final.Country.value_counts().index[:3]
```

```python
df_final.columns
```

```python
ratings=df_final.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

What we can infer from the above df is:
    when the ratings are b/w 1.8 to 2.4, they're considered to be poor.
    when the ratings are b/w 2.5 to 3.4, they're considered to be average.
    when the ratings are b/w 3.5 to 3.9, they're considered to be good.
    when the ratings are above 4.5, they're considered to be excellent.

```python

import seaborn as sns
from matplotlib.pyplot import figure

figure(figsize=(15, 6), dpi=80)

sns.barplot('Aggregate rating', 'Rating count', data=ratings, hue="Rating text",dodge=False, palette=['blue', 'red', 'orange', 'yellow', "green", "green"])
```

Observations:
Max number of ratings are between 2.5 and 3.4.
A large number of people haven't even rated.


```python
sns.countplot(x="Rating text", data=ratings, palette=['blue', 'red', 'orange', 'yellow', "green", "green"])
```
